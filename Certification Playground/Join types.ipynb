```python
%sql
use catalog playground;
SELECT * FROM todos.tasks t
inner join todos.task_status ts on ts.task_id = t.task_id
inner join todos.status s on ts.status_id = s.status_id
order by t.task_id,s.status_id;
```

```python
%sql
use catalog playground;
SELECT * FROM todos.status s
left join todos.task_status ts on s.status_id = ts.status_id;
```

```python
%sql
use catalog playground;
SELECT * FROM todos.tasks t
right join todos.task_status ts on t.task_id = ts.task_id;
```

```python
%sql
use catalog playground;


SELECT * FROM todos.status s
full join todos.task_status ts on s.status_id = ts.status_id
full join todos.tasks t on ts.task_id = t.task_id;
```

```python
%sql
--Easier form of intersect
use catalog playground;
SELECT * FROM todos.status s
semi join todos.task_status ts on s.status_id = ts.status_id;
```

```python
%sql
--easier form of except
use catalog playground;
SELECT * FROM todos.tasks t
anti join todos.task_status ts on t.task_id = ts.task_id;
```

```python
%sql
use catalog playground;
with cte as (
SELECT task_id FROM todos.tasks t
intersect
SELECT task_id from todos.task_status
)
SELECT * FROM cte c inner join todos.tasks t on c.task_id = t.task_id
;
```

```python
%sql
CREATE OR REPLACE PROCEDURE todos.semi_join_task_status_tasks()
LANGUAGE SQL
SQL SECURITY INVOKER
AS
BEGIN
  SELECT t.task_id, t.task_name, t.description, t.assigned_to, t.due_date
  FROM todos.tasks t
  SEMI JOIN todos.task_status ts ON t.task_id = ts.task_id;
END;
```

```python
%sql
CALL todos.semi_join_task_status_tasks();
```
